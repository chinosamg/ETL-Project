```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
sns.set()
plt.gcf().subplots_adjust(bottom=0.15)
plt.tight_layout()
```

```python
Fortune500_2016_file = "./Resources/Fortune500_2016.csv"
Fortune500_2016 = pd.read_csv(Fortune500_2016_file)
Fortune500_2016.dropna(axis=1, how='all') 
```

```python
Fortune500_2017_file = "./Resources/Fortune500_2017.csv"
Fortune500_2017 = pd.read_csv(Fortune500_2017_file)
Fortune500_2017.dropna(axis=1, how='all')

Fortune500_2017 = Fortune500_2017.rename(columns={"id":"ID", "Title":"Company", "Hqlocation":"HQ"})
Fortune500_2017 = Fortune500_2017.drop(columns={"Website", "Industry", "Hqcity", "Hqaddr","Hqzip", "Hqtel", "Ceo-title", "Address", 
                                     "Fullname", "Revchange", "Prftchange", "Assets", "Totshequity", "Ceo", "Hqstate"})
Fortune500_2017.columns
Fortune500_2017                   
```

```python
stock_table = pd.merge(Fortune500_2016, Fortune500_2017, on="Company")
stock_table
```

```python
stock_transformed = stock_table.rename(columns={"Ticker_x" : "Ticker",
       "Rank_x": "Rank_2016",
        "Revenues_x" : "Revenues_2016",
       "Rank_y" : "Rank_2017",
       "Ticker_y" : "Ticker2",
       "Revenues_y" : "Revenues_2017"})
stock_transformed = stock_transformed.drop(columns={"Profits", "Ticker2"})
stock_transformed
```

```python
stock_revenue_compare = stock_transformed.drop(columns={"Rank_2016", "Rank_2017"})
stock_revenue_compare
```

```python
x_axis = np.arange(len(stock_revenue_compare))
tick_locations = [value for value in x_axis]
```

```python
colors = ["blue", "green"]
multi_plot = stock_revenue_compare.plot(kind="bar", figsize=(20,5), color=colors)
plt.xticks(tick_locations, stock_revenue_compare["Ticker"], rotation="horizontal", fontsize=(30))
```

```python
ticker_df = stock_revenue_compare.drop(columns={"Revenues_2016", "Revenues_2017"})
ticker_df = ticker_df.set_index("Ticker")
ticker_df
```
