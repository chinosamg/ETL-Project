```python
import pandas as pd
from sqlalchemy import create_engine
import os
import pymysql
pymysql.install_as_MySQLdb()
```

```python
Fortune500_2017_file = "./Resources/Fortune500_2017.csv"
Fortune500_2017 = pd.read_csv(Fortune500_2017_file)
Fortune500_2017.dropna(axis=1, how='all')
```

```python
Fortune500_2017 = Fortune500_2017.rename(columns={"id":"ID", "Title":"Company", "Hqlocation":"HQ"})
Fortune500_2017 = Fortune500_2017.drop(columns={"Website", "Industry", "Hqcity", "Hqaddr","Hqzip", "Hqtel", "Ceo-title", "Address", 
                                     "Fullname", "Revchange", "Prftchange", "Assets", "Totshequity", "Ceo", "Hqstate"})
Fortune500_2017.columns
                             
```

```python
Fortune500_2017.head(10)
```

```python
Fortune500_2017_cols = ["Rank", "Company", "Ticker", "Revenues"]
Fortune500_2017_transformed= Fortune500_2017[Fortune500_2017_cols].copy()

Fortune500_2017_transformed.drop_duplicates({"Rank", "Company"}, inplace=True)
Fortune500_2017_transformed.set_index("Rank", inplace=True)

Fortune500_2017_transformed.head()
```

```python
Fortune500_2016_file = "./Resources/Fortune500_2016.csv"
Fortune500_2016 = pd.read_csv(Fortune500_2016_file)
Fortune500_2016.dropna(axis=1, how='all')
```

```python
Fortune500_2016 = Fortune500_2016.rename(columns={"SYMBOL":"Ticker", "COMPANY":"Company", "REVENUES($M)":"Revenues", "RANK":"Rank"})
Fortune500_2016 = Fortune500_2016.drop(columns={"WEBSITE"})
Fortune500_2016.columns                    
```

```python
Fortune500_2016.head(10)
```

```python
Fortune500_2016_cols = ["Rank", "Company", "Ticker", "Revenues"]
Fortune500_2016_transformed= Fortune500_2016[Fortune500_2016_cols].copy()

Fortune500_2016_transformed.drop_duplicates({"Rank", "Company"}, inplace=True)
Fortune500_2016_transformed.set_index("Rank", inplace=True)

Fortune500_2016_transformed.head()
```

```python
connection_string = "root:@localhost/ETL_Project_Fortune500"
engine = create_engine(f'mysql://{connection_string}')
```

```python
# Confirm tables
engine.table_names()
```

```python
Fortune500_2017_transformed.to_sql(name='fortune500_2017', con=engine, if_exists='replace', index=True)
```

```python
Fortune500_2016_transformed.to_sql(name='fortune500_2016', con=engine, if_exists='replace', index=True)
```
