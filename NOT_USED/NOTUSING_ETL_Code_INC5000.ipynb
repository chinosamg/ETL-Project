```python
import pandas as pd
from sqlalchemy import create_engine
import pymysql
pymysql.install_as_MySQLdb()
```

```python
inc2016_file = "./Resources/inc5000_2016.csv"
inc2016_df = pd.read_csv(inc2016_file)
inc2016_df.dropna(axis=1, how='all')
```

```python
inc2016_df = inc2016_df.rename(columns={"id":"ID", "rank":"Rank", "workers":"Employees", "company":"Company2016",
                                       "url":"Website", "city":"City", "revenue": "Revenue",
                                       "yrs_on_list":"Years On List", "state_l":"State"})
inc2016_df = inc2016_df.drop(columns={"growth", "ifmid", "ifiid", "metro", "industry","state_s"})
inc2016_df.columns
```

```python
inc2016_df.head()
```

```python
inc2018_file = "./Resources/inc5000-2018.csv"
inc2018_df = pd.read_csv(inc2018_file)
inc2018_df.dropna(axis=1, how='all')
```

```python
inc2018_df = inc2018_df.rename(columns={"_ - id":"ID", "_ - rank":"Rank", "_ - company":"Company2018",
                                       "_ - url":"Website", "_ - city":"City", "_ - revenue": "Revenue",
                                       "_ - yrs_on_list":"Years On List", "_ - workers":"Employees","_ - state_l":"State"})
inc2018_df = inc2018_df.drop(columns={"_ - industry", "_ - ifmid", "_ - ifiid", "_ - website", "_ - zipcode", "_ - founded",
                                     "_ - state_s", "_ - latitude", "_ - metrocode", "_ - longitude", "_ - previous_workers", 
                                     "_ - metro", "_ - partner_lists - partner_lists", "_ - growth"})
inc2018_df.columns
```

```python
inc2018_df.head()
```

```python
inc2016_cols = ["ID", "Company2016", "Rank", "Revenue"]
inc2016_transformed= inc2016_df[inc2016_cols].copy()

inc2016_transformed.drop_duplicates("ID", inplace=True)
inc2016_transformed.set_index("ID", inplace=True)

inc2016_transformed.head()
```

```python
inc2018_cols = ["ID", "Rank", "Company2018", "Revenue"]
inc2018_transformed= inc2018_df[inc2018_cols].copy()

inc2018_transformed.drop_duplicates("ID", inplace=True)
inc2018_transformed.set_index("ID", inplace=True)

inc2018_transformed.head()
```

```python
connection_string = "root:@localhost/ETL_Project"
engine = create_engine(f'mysql://{connection_string}')
```

```python
# Confirm tables
engine.table_names()
```

```python
inc2016_transformed.to_sql(name='inc2016', con=engine, if_exists='append', index=True, 'utf-8', 'replace')
```

```python
inc2018_transformed.to_sql(name='inc2018', con=engine, if_exists='append', index=True)
```

```python
inc2016_2018 = pd.merge(inc2016_df, inc2018_df, on="ID", how="left")
```

```python
inc2016_2018 = inc2016_2018.rename(columns={"Rank_x":"Rank2016", "Employees_x":"Employees2016", "Website_x":"Website2016",
                                           "State_x":"State2016", "City_x":"City2016", "Revenue_x":"Revenue2016",
                                           "Years On List_x":"Years On List2016", "Website_y":"Website 2018",
                                           "Rank_y":"Rank2018", "City_y":"City2018", "Employees_y":"Employees2018",
                                           "State_y":"State2018", "Revenue_y":"Revenue2018", "Years On List_y":"Years On List 2018"})
inc2016_2018
```

```python
inc2016_2018_cols = ["ID", "Company2016","Company2018", "Rank2016"]
inc2016_2018_transformed= inc2016_2018[inc2016_2018_cols].copy()

inc2016_2018_transformed.drop_duplicates("ID", inplace=True)
inc2016_2018_transformed.set_index("ID", inplace=True)

inc2016_2018_transformed.head()
```

```python
inc2018_transformed.to_sql(name='inc2018', con=engine, if_exists='append', index=True)
```
