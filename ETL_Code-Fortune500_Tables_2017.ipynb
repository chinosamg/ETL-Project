```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
sns.set()
plt.gcf().subplots_adjust(bottom=0.15)
plt.tight_layout()
```

```python
Fortune10_2017_file = "./Resources/Fortune10_2017.csv"
Fortune10_2017 = pd.read_csv(Fortune10_2017_file)
Fortune10_2017.dropna(axis=1, how='all')
```

```python
x_axis = np.arange(len(Fortune10_2017))
tick_locations = [value for value in x_axis]
```

```python
plt.figure(figsize=(35,15))
colors = ["green", "blue"]

plt.bar(x_axis, Fortune10_2017["Revenues"], alpha=.9, align="center", color=colors)
plt.xticks(tick_locations, Fortune10_2017["Company"], rotation="vertical", fontsize=(30))
plt.savefig("Resources/Top10_2017_Revenue.png")
```
